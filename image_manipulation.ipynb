```python
from PIL import Image 
import cv2
import numpy as np
from pytesseract import image_to_string 
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
#Read
img = cv2.imread('/content/drive/MyDrive/Internship2/Day07/dog.png')
#Show
cv2_imshow(img)
#Gray,Blur,Canny
kernel=np.ones((5,5),np.uint8)
imGray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
imBlur=cv2.GaussianBlur(imGray,(7,7),0)
imCanny=cv2.Canny(img,100,100)
cv2_imshow(imGray)
cv2_imshow(imBlur)
print("\n \n")
cv2_imshow(imCanny)
#Dialation,Erosion
kernel = np.ones((5,5), np.uint8)  
img_erosion = cv2.erode(img, kernel, iterations=1)  
img_dilation = cv2.dilate(img, kernel, iterations=1)  
cv2_imshow(img_erosion)  
cv2_imshow(img_dilation)   
#Resize
# 1) Smaller Than Original
imgResize=cv2.resize(img,(100,110))
cv2_imshow(imgResize)
# 2) Upto to screen size 
imgScreen = cv2.resize(img,(1600,800))
cv2.imwrite("/content/drive/MyDrive/Internship2/Day07/temp/Screensize_dog.png",imgScreen)
cv2_imshow(imgScreen)
cv2.waitKey(0)

#Crop Image
imgCropped=img[30:700,110:500]
cv2_imshow(imgCropped)
#Insertion of shapes
rectangle = cv2.rectangle(img,(107,0),(200,150),(0,100,0),3)
circle = cv2.circle(img,(137,60), 60, (0,0,100), 2)
cv2_imshow(rectangle)
#Joining 2 images
img = cv2.imread('/content/drive/MyDrive/Internship2/Day07/dog.png')
img1=cv2.imread("/content/drive/MyDrive/Internship2/Day07/tree.png")
imgHor=np.hstack((img,img1))
cv2_imshow(imgHor)
```
