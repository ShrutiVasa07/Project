```python
from google.colab.patches import cv2_imshow
import numpy as np
import cv2
faceCascade= cv2.CascadeClassifier("/content/drive/MyDrive/Day-4 (5th June, 2023) Material/haarcascades/haarcascade_frontalface_alt.xml")
img = cv2.imread('/content/drive/MyDrive/Internship2/Day07/Group.jpeg.jpg')
imgGray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

faces = faceCascade.detectMultiScale(imgGray,1.00465,4)

for (x,y,w,h) in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)

cv2_imshow(img)
cv2.imwrite("/content/drive/MyDrive/Internship2/Day07/temp/Detected.jpg",img)
cv2.waitKey(0)
```
